# **NET PROMOTER SCORE**
1 TO 6 --> Represents DETRACTORS

7 TO 8 --> Represents PASSIVES

9 TO 10 --> Represents PROMOTERS

Calculate NPS

**NPS = % PROMOTERS  -  % DETRACTORS**

```python
print(scores)
```

```python
print(f"Minimum Survey Rating  = {scores.min()}")
print("Maximum Survey Rating  =",scores.max())
print(f"Mean Survey Rating  = {scores.mean()}")

from scipy import stats
print(f"Mode Survey Rating  = {stats.mode(scores)}")
```

```python
# Masking only Promoter Values
promoters = scores[scores > 8]
print(promoters)
```

```python
# Count Promoters
Count_Promoters = len(promoters)
print(Count_Promoters)
```

```python
# Masking only Detractors Values
detractors = scores[scores < 6]
print(detractors)
```

```python
# Count detractors
Count_detractors = len(detractors)
print(Count_detractors)
```

```python
# Count total survey Responses
total_Respondents = len(scores)
print(total_Respondents)
```

```python
# Calculate Percentage of Promoters
percent_promoters = (Count_Promoters * 100) / total_Respondents
print(percent_promoters)
```

```python
# Calculate Percentage of Detractors
percent_detractors  = (Count_detractors * 100) / total_Respondents
print(percent_detractors)
```

**Calculating Net Promoters Score**


```python
NPS = (percent_promoters  -  percent_detractors)
print(NPS)
```
